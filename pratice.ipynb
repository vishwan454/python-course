```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales_trend = [700, 900, 800, 1300, 550, 1700, 1900, 2100, 1850, 1900, 1500, 2000]
plt.figure(figsize=(15,6))
plt.plot(months, sales_trend, marker="o", linestyle="-", color="b")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Hypothetical Sales Trend Over 12 Months')
plt.show()
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
product_A_sales = [700, 900, 800, 1300, 550, 1700, 1900, 2100, 1850, 1900, 1500, 2000]
product_B_sales= sorted(product_A_sales)
plt.figure(figsize=(15,6))
plt.plot(months, product_A_sales, marker="o", linestyle="dashed", color='r')
plt.plot(months, product_B_sales, marker="*", linestyle="-", color="g")

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Hypothetical Sales Trend Over 12 Months')
plt.legend(['Product A', 'Product B']) #adds legend for respective data
plt.grid() #adds grid
plt.show()
```

```python
aqi_df = pd.read_csv("air_quality_no2_long.csv")
aqi_df.head()
```

```python
aqi_df = aqi_df.rename(columns={"date.utc": "datetime"})
aqi_df.head()
```

```python
aqi_df["datetime"] = pd.to_datetime(aqi_df["datetime"])
aqi_df.dtypes
```

```python
aqi_df["datetime"].min()
```

```python
aqi_df["month"] =aqi_df["datetime"].dt.month
aqi_df.head()
```

```python
aqi_df["hour"] = aqi_df["datetime"].dt.day
hour_agg = aqi_df.groupby("hour")[["value"]].agg("mean")
hour_agg.head(9)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(hour_agg.index, hour_agg["value"], marker="o", linestyle="-", color="g")
plt.grid()
plt.show()
```
